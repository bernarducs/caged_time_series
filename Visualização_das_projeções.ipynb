```
import pandas as pd
```

```
# viz
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
sns.set_context("talk")
```

```
file_path = "/content/drive/My Drive/Colab Notebooks/emprego/bases/"
file_names = ("saldo_pe_pred.csv", "saldo_ce_pred.csv", "saldo_ba_pred.csv")
file_names_comercio = ("saldo_pe_comercio_pred.csv", "saldo_ce_comercio_pred.csv", "saldo_ba_comercio_pred.csv")
file_names_servicos = ("saldo_pe_servicos_pred.csv", "saldo_ce_servicos_pred.csv", "saldo_ba_servicos_pred.csv")
```

```
df_list = list()
for file_name in file_names:
  uf = file_name.split("_")[1]
  name_df = f"{uf}"
  name_df = pd.read_csv(file_path+file_name, 
                        index_col=['data'], names=['data', name_df], skiprows=1)
  df_list.append(name_df)
```

```
df = pd.concat(df_list, axis=1)
df.index = pd.to_datetime(df.index)
```

```
fig, ax = plt.subplots(figsize=(16, 6))
plt.figure(figsize=(12, 6))

df.ba.plot(ax=ax, style='-o', linewidth=2, alpha=0.5)
df.ce.plot(ax=ax, style='-o', linewidth=2, alpha=0.5)
df.pe.plot(ax=ax, style='g-o', linewidth=4, alpha=0.7)

ax.set(xlabel="Ano",
       ylabel="Saldo de Vagas",
       title="Saldo de vagas - GERAL",
       xlim=["2010-10-01", "2021-02-01"])

ax.legend(loc='lower left');
```
